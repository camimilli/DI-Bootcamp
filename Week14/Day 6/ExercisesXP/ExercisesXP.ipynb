## Exercise 1 - Use The Openpyxl Module To Create A New Workbook And Add Values/Formulas To It

```python
pip install openpyxl xlrd
```

```python
from openpyxl import Workbook 
from openpyxl.utils import get_column_letter

# Create workbook 
wb = Workbook()
ws = wb.active
ws.title = "Calculator"

# Create data and append to worksheet
data = [
    ["First number ==>", 2],
    ["Second number ==>", 8]
    ]

for row in data:
    ws.append(row)

#Auto-adjust Column Widths
for column in ws.columns:
    max_length = max(len(str(cell.value)) if cell.value is not None else 0 for cell in column)
    adjusted_width = max_length + 2 # add some "padding" so the text is not very close to the cell borders
    ws.column_dimensions[get_column_letter(column[0].column)].width = adjusted_width

# Add multiplication formula 
ws['B3'] = f"=B1*B2"

# Save workbook
wb.save("Calculator.xlsx")

# Close workbook
wb.close()
```

## Exercise 2 - Open an Excel workbook and read down a list of plants finding the ones not in stock

```python
from openpyxl import load_workbook

# Load Workbook 
wb = load_workbook('Plants.xlsx')
ws = wb.active 

# Get worksheet max rows 
max_row = ws.max_row

# Iterate through the plant rows to check which plant is out of stock 
for row in range (2, max_row + 1):
    in_stock = ws['H' + str(row)].value 
    if in_stock == "No":
        print(f"{ws['A' + str(row)].value} is not on stock")

wb.close()
```

## Exercise 3 - Load Excel File And Write The Data

```python
import pandas as pd 

# Read excel file into DataFrame
df = pd.read_excel('data.xlsx')

# Inspect data
print(df.head())
df.info()
```

```python
# Filter data where sales is > 1000 
df = df[df['Sales'] > 1000]

# Write filtered data back to Excel file 
with pd.ExcelWriter("data.xlsx", engine='openpyxl') as writer:
    df.to_excel(writer, sheet_name='Sheet1', index=False)
```

## Exercise 4 - Load Excel File, Plot Charts And Manipulate Data

```python
# Load excel file into DataFrame
df = pd.read_excel('productSales.xlsx')

# Inspect data
print(df.head())
df.info()
```

```python
# Group by product + sum the sales for each product 
df = df.groupby('product')['sales'].sum().reset_index()

# Export back into new Excel file 
df.to_excel("sales_report.xlsx", sheet_name='Sales', index=False)
```
