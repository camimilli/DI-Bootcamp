## Daily Challenge - Excel Gradebook Creation

```python
# Import libraries
from openpyxl import Workbook
from openpyxl.styles import Font 
from openpyxl.utils import get_column_letter
```

```python
# Data to append into Excel file 

data = {
    "Joe": {
        "math": 65,
        "science": 78,
        "english": 98,
        "gym": 89
    },
    "Bill": {
        "math": 55,
        "science": 72,
        "english": 87,
        "gym": 95
    },
    "Tim": {
        "math": 100,
        "science": 45,
        "english": 75,
        "gym": 92
    },
    "Sally": {
        "math": 30,
        "science": 25,
        "english": 45,
        "gym": 100
    },
    "Jane": {
        "math": 100,
        "science": 100,
        "english": 100,
        "gym": 60
    }
}
```

```python
# Create workbook
wb = Workbook()
ws = wb.active
ws.title = "Grades"

# Add headings 
headings = ["Name"] + list(data['Joe'].keys())
ws.append(headings)

# Add a row for each student's grades 
for person in data:
    ws.append([person] + list(data[person].values()))

# Add average grade at end of each column
max_column = ws.max_column # Get max column number

for col in range(2, max_column + 1):
    col = get_column_letter(col)
    ws[col + '7'] = f"=AVERAGE({col + '2'}:{col + '6'})"

# Apply bold and colored formatting to the header row 
for col in range(1, max_column + 1):
    col = get_column_letter(col)
    ws[col + "1"].font = Font(bold=True, color="00008000")

wb.save("Grades.xlsx")
```
